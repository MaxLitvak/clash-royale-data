```python
from sklearn.cluster import KMeans
```

```python
independent = data.drop('winRateDif',axis=1)
```

```python
kmeans = KMeans(n_clusters=3)
```

```python
kmeans.fit(independent)
```

```python
np_data = independent.values
labels = []
for row in np_data:
    vals = row.reshape(1, -1)
    label = kmeans.predict(vals)
    label = label[0]
    labels.append(label)
data['labels'] = labels
```

```python
data
```

```python
sns.set_style('whitegrid')
sns.lmplot('damageDif', 'winRateDif', data=data, hue='labels', palette='coolwarm', size=8, aspect=1, fit_reg=False)
```

This graph is based on clash royale. It is a game where two players each have a deck with 8 cards. Each player also has three towers. Whichever player destroys more towers regardless of the total damage wins. Here the x-axis is the damage difference in the two players' towers. Some values are negative because a player can win by destroying more of the opponents towers even if the opponent did more accumlated damage. The y-axis is the win rate difference in the 8-cards that each player used. To calculate win rate difference, I addeded up the win rates of all of the winners' cards and then subtrated the sum of the win rates of the loser. This graph shows that the game is purely based on skill because there is no trend.
